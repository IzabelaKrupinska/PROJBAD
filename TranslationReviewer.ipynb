```
import matplotlib.pyplot as plt
import os
import pickle
import pandas as pd
import random


from google.colab import drive
drive.mount('/content/drive', force_remount=True)

base_dir ='/content/drive/MyDrive/MB/'
images_dir = '/content/drive/My Drive/Flickr8k_Dataset/Flicker8k_Dataset/'
translation_dir = '/content/drive/My Drive/Translations/'
```

```
def plot_img(filename):
  image = plt.imread(images_dir + filename)
  plt.imshow(image)
  plt.show()
```

```
# Loading Descriptions
descriptions = pickle.load(open(base_dir + "descriptions_text.pkl", "rb"))
print('Loaded %d descriptions' % len(descriptions))
```

```
translations = {}
for entry in os.listdir(translation_dir):
  entry_path = os.path.join(translation_dir, entry)
  if os.path.isfile(entry_path):
    loaded_translations = pickle.load(open(entry_path, 'rb'))
    translations.update(loaded_translations)

#sanity check
print(len(translations) == len(descriptions))
all([len(v) == 5 for v in translations.values()])
```
